#### **Getting all the columns of the movie dataset**

<h1><font size=6>Can we find the movies that have more than one genre?</font></h1>

As discussed earlier a particular movie can have more than one genre, so to answer the above problem we can use the below steps:


- The basic approach to find out this information will be to add a new column to the movie dataframe that will contain the summation of 1's from the genre columns along the row for a particular movie
- After this we can find out the number of movies that have more than one genre by simply picking up the rows where the added new column is having a value of more than 1

So, let us create a new column named `Number of Genres` and add it to the dataframe.

```python
# Creating a new column 'Number of Genres' and adding all the genre counts for the particular movie
movie_df['Number of Genres'] = movie_df[movie_genres].sum(axis=1)
```

Now let us print the head of the movie column and see whether the new column has been added or not.

```python
# Fetching the first 5 rows of the movie dataset
movie_df.head()
```

- *As we can see a new column named `Number of Genres` has been added at the end of the dataframe*

Now let us pick up the rows where `Number of Genres`>1 so that we can find the number of movies with more than 1 genre

```python
# Finding the number of movies that have more than 1 genre
movies_with_multiple_genres = movie_df[movie_df['Number of Genres'] > 1]
```

Let us print the shape of this dataframe so that we know the number of movies that are having more than one genre.

```python
# Printing the number of movies with more than 1 genre
print(f"Number of movies with more than one genre: {movies_with_multiple_genres.shape[0]}")
```

<h1><font size=6>Can we find the genres that are most liked by the users?</font></h1>


*As we have already seen that the information regarding the ratings is present in the ratings dataset, however, the genre columns are present in the movie dataset. The only column connecting this two datasets is the movie id column. We can `merge` both the datasets on the basis of common variable that is `movie id`.*

**Merge/Join / Concatenation**<br>
Pandas provide various facilities for easily combining together Series or DataFrame with various kinds of set logic for the indexes and relational algebra functionality in the case of join / merge-type operations.<br> <br>Here are the different types of the Joins:<br>

**Inner Join**: Returns records that have matching values in both tables<br>

inner join.png

**Left Join**: Returns all records from the left table, and the matched records from the right table<br>

left.png

**Right Join**: Returns all records from the right table, and the matched records from the left table<br>

right.png

**Outer/Full Join**: Returns all records when there is a match in either left or right table

outer.png

We'll now join the movie data and ratings data on `movie id` to find the genres that are most liked by users.

```python
# Merged movie and ratings data based on the 'movie id'
merged_df = pd.merge(movie_df, ratings_df, on='movie id')
merged_df
```

```python
# Getting the shape of the merged dataframe
print("Shape of the merged dataframe:", merged_df.shape)
```

In order to find the genres which are most liked by the users, we can find the average ratings of each of the genre and then compare among them.
<br>
For this we can use the `genres` list

We will run a loop across the merged dataframe for the columns that are present in the genres list. Whenever a particular genre is found to be have '1' value in its column, the rating corresponding to that '1' value will be extracted and in this end we will calculate the average value of all this extracted ratings for a particular genre.

```python
# Finding the movie genres' average ratings
genre_avg_ratings = {}

for genre in movie_genres:
    genre_avg_ratings[genre] = merged_df[merged_df[genre] == 1]['rating'].mean()

genre_avg_ratings = pd.Series(genre_avg_ratings).sort_values(ascending=False)

print("Average ratings for each movie genre:")
genre_avg_ratings
```

Printing only the top three most liked movie genres based on the user reviews

```python
# Top three movie genres based on user ratings
genre_avg_ratings.head(3)
```

- `Film-Noir`, `War`, and `Drama` are the most liked genres by users, having the highest average ratings among all genres.

**Groupby Function**<br>
Pandas’ *GroupBy* is a powerful and versatile function in Python. It allows you to split your data into separate groups to perform computations for better analysis.*GroupBy* allows us to group our data based on different features and get a more accurate idea about your data.

<h1><font size=6>Can we extract some relationship between demographic details of the users and ratings for the movies?</font></h1>

****



All the datasets can be used to identify the demography of the users. Therefore, we will merge all the datasets together and perform the analysis

Since we have already merged datasets `rating` and `movie` in a variable `merged_df`, therefore we can merge the dataset `user` with this dataset

```python
# Merge the user dataset with the already merged data of ratings and movie dataset
all_merged_df = pd.merge(merged_df, user_df, on='user id', how = 'left')
all_merged_df
```

**Let's check on various attributes of the users in order to know the preferences for the movies and their genres**

```python
# Average rating by each gender
all_merged_df.groupby(['gender'])['rating'].mean()
```

- *The average rating given by Males and Females is almost equal with a value of around 3.53*

```python
# Average rating by gender per genre count
all_merged_df.groupby(['gender', 'Number of Genres'])['rating'].mean().unstack()
```

```python
# Average rating by gender for each movie and overall average rating
movie_gender_avg = all_merged_df.pivot_table(
    index='movie title',
    columns='gender',
    values='rating',
    aggfunc='mean'
)

movie_gender_avg['Overall'] = movie_gender_avg.mean(axis=1)
movie_gender_avg.sort_values(by='Overall', ascending=False, inplace=True)
movie_gender_avg
```

**We can see that few movies are not rated by males and few movies are not rated by females. Let's get the list of male only and female only rated movies**

```python
# Movies that are rated by males
male_movies = all_merged_df[all_merged_df['gender'] == 'M']['movie title'].unique()

# Movies that are rated by females
female_movies = all_merged_df[all_merged_df['gender'] == 'F']['movie title'].unique()
```

```python
# Movies that are rated by males but not by females
male_only_movies = set(male_movies) - set(female_movies)

print("Movies rated by males but not by females:", len(male_only_movies))
male_only_movies
```

- We can see that `141` movies are rated only by males.

```python
# Movies that are rated by females but not by males
female_only_movies = set(female_movies) - set(male_movies)

print("\nMovies rated by females but not by males:", len(female_only_movies))
female_only_movies
```

- We can see that `74` movies are rated only by females.

### **2. Bivariate Analysis**

**Age and Occupation comparison - (Numerical vs Categorical)**

```python
# Boxplot for Age vs Occupation
plt.figure(figsize=(12, 6))
sns.boxplot(x='occupation', y='age', data=user_df)
plt.xticks(rotation=45)
plt.xlabel('Occupation')
plt.ylabel('Age')
plt.title('Age Distribution by Occupation')
plt.show()
```

**Number of Ratings vs Average Rating - (Numerical vs Numerical)**

```python
# Number of Ratings vs Average Ratings - Regression Plot
movie_stats = ratings_df.groupby('movie id')['rating'].agg(['mean', 'count'])

plt.figure(figsize=(8,6))
sns.regplot(data=movie_stats, x='count', y='mean', line_kws={'color': 'red'})
plt.xlabel('Number of Ratings')
plt.ylabel('Average Rating')
plt.title('Number of Ratings vs Average Rating')
plt.show()
```

- *Most movies receive only a small number of ratings, indicating that only a **few movies are widely watched and reviewed**.*

- *Movies with a larger number of ratings generally have **higher average ratings**, suggesting that **popular movies are often well received by viewers**.*

- *Low-rated movies rarely accumulate many ratings, whereas highly rated movies are more likely to attract a large number of viewers and ratings.*


### **3. Multivariate Analysis**

**Movie Genre x Average Ratings x Gender**

```python
# Average rating by gender for each movie genre
genre_gender = {}

for genre in movie_genres:
    genre_gender[genre] = all_merged_df[all_merged_df[genre] == 1].groupby('gender')['rating'].mean()

# Create dataframe and visualize using heatmap
genre_gender_df = pd.DataFrame(genre_gender).T

plt.figure(figsize=(8, 8))
sns.heatmap(genre_gender_df, annot=True, cmap='YlGnBu', fmt='.2f')
plt.title('Average Ratings by Genre and Gender')
plt.show()
```

- *Male users generally rate movies slightly higher than female users in many genres. For example, **Film-Noir** is rated `3.97` by males vs `3.74` by females, **Western** is `3.64 vs 3.51` and **Crime** is `3.65 vs 3.56`.*

- *Genres like **Film-Noir** and **War** are the highest-rated genres for both genders. Film-Noir has the highest average ratings (**3.74 for females, 3.97 for males**), while War also receives strong ratings (**3.78 for females, 3.83 for males**), showing broad appeal across genders.*

- *Genres like **Fantasy** and **Horror** genres receive the lowest ratings overall. Fantasy has the lowest averages (**3.20 for females, 3.22 for males**), followed by Horror (**3.26 for females, 3.30 for males**) indicating comparatively lower audience satisfaction.*
